### Question 2 Removing NAN and outliers from engine_displacement column

```python
data['engine_displacement'] = data['engine_displacement'].replace(['none', 'nothing', 'NaN'], ['NaN', 'NaN', 'NaN'])
```

```python
data['engine_displacement'] = data['engine_displacement'].astype(float)
```

```python
data['engine_displacement'] = data['engine_displacement'].fillna((data['engine_displacement'].median()))
```

### I have removed NaN values and made the median a replacement for the NaN values

```python
data['engine_displacement'].value_counts()[:5]
```

```python
low = .01
high = .99
quant_df = data['engine_displacement'].astype(float).quantile([low, high])
quant_df_low = data['engine_displacement'].astype(float).quantile([high])
print(list(quant_df_low)[-1])
```

### I set outliers to be numbers lower than the 1th and numbers greater than the 99th percentile which is 1.5 and 5.6 respectively

```python
data['engine_displacement'] = [x if(x >=1.5 <=5.6) else 1.5 for x in list(data['engine_displacement'])]
```

### Question 3 Create columns based on the contents of the tags column (for example a new column might be a (logical) under_warranty with values FALSE/TRUE.


```python
def Tag_Bool(data):
    if data['tags'] == '{under_warranty}':
        return True
    return False
```

### Question 10 Create a column named recency where you calculate the number of days between last_seen_at and today’s date.


```python
date1 = pd.to_datetime(data.last_seen_at)
```

```python
def recency():
    return [pd.Timestamp.now(tz='UTC') -pd.Timestamp(i) for i in date1]
```

```python
data['recency'] = recency()
data['recency']
```
